```python
#creating probability density to determine caustic, used in fig 7

#takes in particle positions (angle and radius) f and r in time, list of relevant points (taking out high density at boudnary), bounds of timesteps I am interested in, and size of grid
def setgrid(f, r, listit, minind, maxind, gridsize):

    # Compute x and y lists using polar coordinates
    xlist = r * np.cos(f)
    ylist = r * np.sin(f)
    
    # Get the region of interest for x and y
    x_min = np.min(xlist[listit][minind:maxind])
    x_max = np.max(xlist[listit][minind:maxind])
    y_min = np.min(ylist[listit][minind:maxind])
    y_max = np.max(ylist[listit][minind:maxind])
    
    # Create grid points
    x = np.linspace(x_min, x_max, gridsize)
    y = np.linspace(y_min, y_max, gridsize)
    
    # Generate grid edges
    x_edges = np.digitize(xlist[listit][minind:maxind], x) - 1
    y_edges = np.digitize(ylist[listit][minind:maxind], y) - 1
    
    # Ensure values fall within valid bins
    x_edges = np.clip(x_edges, 0, len(x) - 2)
    y_edges = np.clip(y_edges, 0, len(y) - 2)
    
    # Compute the 2D histogram to count points in each grid cell
    histogram, _, _ = np.histogram2d(
        xlist[listit][minind:maxind],
        ylist[listit][minind:maxind],
        bins=[x, y]
    )
    
    # Prepare the output
    xset, yset = np.meshgrid(x[:-1], y[:-1], indexing='ij')
    xset = xset.ravel()
    yset = yset.ravel()
    zset = histogram.ravel()
    
    return xset.tolist(), yset.tolist(), zset.tolist()


#X, Y and Z are determined from the function above

zeroind= np.where((np.array(Z) >0) & (np.array(Z) <1000))
# Create points from x and y sets
points = np.column_stack((np.array(X)[zeroind], np.array(Y)[zeroind]))

# Compute the alpha shape
# alpha = 2
# alpha = 5
alpha =5
hull = alphashape.alphashape(points, alpha)

# Check if hull is empty or not
if hull.is_empty:
    print("The alpha shape is empty. Adjust the alpha value.")
else:
#     a = 327
#     b =-36
#     c = -115
#     d = -1
#     a = 350
#     b =-23
    a = 207
    b = -138

        # Extract the hull points
    hull_points51 = np.array(hull.exterior.coords)[a:b]# Get exterior coordinates of the alpha shape, the coordinates are adjusted by hand


plt.scatter(hull_points51[:, 0], hull_points51[:, 1], label='Alpha Shape')
plt.scatter(np.array(X)[zeroind], np.array(Y)[zeroind], c= np.array(Z)[zeroind], s = .1)


plt.scatter(boundx, boundy,s = 1)


# Function to calculate distance between two points
def distance(p1, p2):
    return np.linalg.norm(p2 - p1)

# Function to create evenly spaced points along each edge
def generate_evenly_spaced_points(hull_points, num_points):
    total_length = sum(distance(hull_points[i], hull_points[i + 1])
                       for i in range(len(hull_points) - 1))  # No wrapping around
    step_size = total_length / num_points
    
    evenly_spaced_points = [hull_points[0]]  # Start with the first point
    current_point = hull_points[0]
    current_idx = 0  # Index of the current vertex
    
    # Remaining distance on the current edge to move to the next point
    remaining_distance = step_size

    while len(evenly_spaced_points) < num_points:
        if current_idx + 1 < len(hull_points):  # Avoid wrapping around
            next_point = hull_points[current_idx + 1]  # Get the next vertex
            dist_to_next = distance(current_point, next_point)  # Distance to next vertex
            
            # If the next vertex is farther than or exactly at the step size, add the new point
            if dist_to_next >= remaining_distance:
                direction = (next_point - current_point) / dist_to_next  # Normalize direction
                new_point = current_point + direction * remaining_distance  # Move by step_size
                evenly_spaced_points.append(new_point)  # Add new point
                current_point = new_point  # Update current point
                remaining_distance = step_size  # Reset the remaining distance to step_size
            else:
                # Move to the next vertex and adjust remaining distance accordingly
                remaining_distance -= dist_to_next
                current_idx += 1
                current_point = next_point
        else:
            break  # Exit if we've reached the end of the hull points

    return np.array(evenly_spaced_points)

def normals_to_hull(hull_points,num_points, zset):

    zeroind = np.where((np.array(zset) >0) &(np.array(zset) <1000))
    evenly_spaced_hull_points = generate_evenly_spaced_points(hull_points, num_points)

    h =evenly_spaced_hull_points
    # h=hull_points
    # Calculate normal vectors
    normals = []
    for i in range(1,len(h)):  # Loop through edges
        # Get current and next point (do not wrap around here)
        p0 = h[i-1]
        p1 = h[i]
    #     p2 = h[i + 1]

        # Compute the tangent vector
        tangent = p1 - p0
        # Avoid normalizing a zero vector
        if np.linalg.norm(tangent) > 0:
            tangent = tangent / np.linalg.norm(tangent)

            # Compute the normal vector (90 degrees rotation)
            normal = np.array([tangent[1], -tangent[0]])
            normals.append(normal)
    
    return normals, h,zeroind


def y(t,i,n,h): 
    return n[i][1]*t+h[:, 1][0:-1][i]
def x(t,i,n,h):
    return n[i][0]*t+h[:, 0][0:-1][i]
def ztotcal(X, Y, Z, N, H):
    gridlen= 300
    xgrid = np.linspace(min(boundx), max(boundx),gridlen)
    ygrid = np.linspace(min(boundy), max(boundy),gridlen)
    ztot = []
    # t = np.linspace(0, 1)

    for i in range(len(N)):
        zval =[]
    # i =65
    # 
        for j in range(len(t)-1):
        #     j = len(t)-1
            tchoice= t[j]
            if x(tchoice, i, N,H)>=max(xgrid)-0.02 and y(tchoice,i,N,H) >= max(ygrid)-0.02:
                print('hi')
                break
            else:
                xloc = np.searchsorted(X, x(tchoice, i,N,H))
        #         print(xloc,x(tchoice, i), max(xset))
                xlocend = xloc + gridlen-2
                yloc = xloc+ np.searchsorted(Y[xloc: xlocend], y(tchoice, i, N, H))-1
#                 print(xloc, xlocend)
                plt.scatter(X[yloc], Y[yloc], c = 'b', s =5)
#                  print(xloc, yloc,len(xgrid)*(xloc-1)+yloc)
#                     print(yloc, zloc)
    
                zloc = Z[yloc]
                zval = zval+[zloc]
#         plt.show()
        ztot = ztot+[zval]
    return ztot


def criticalexp(ztot):
    listex = []
    listshift =[]
    listproduct = []
    listscale = []
    b_fitted = 1


    for i in range(3,len(ztot)):
        indcut = []

        rescaledztot = (ztot[i]/(np.sum(ztot[i][np.argmax(ztot[i]): -1])))[0:]
        
        
        listind = np.where(rescaledztot<0.002)
        print(np.delete(rescaledztot, listind)[-3])
        listind = np.where(rescaledztot<np.delete(rescaledztot, listind)[-3])
        
        if len(listind[0])==len(ztot[i]):
            print(i)
            continue
        else:
            for k in listind[0]:

                if k ==0 or k == len(ztot[i])-1:
                    continue
                else:

                    indcut = indcut+[k]

            zi = np.delete(rescaledztot, indcut)[2:int(len(rescaledztot)*0.58)]
            
            # Example data (x and y)
            
            x_data = np.arange(len(zi[np.argmax(zi): -1]))[1:-1]
            y_data = zi[np.argmax(zi): -1][1:-1]
            
            zi2 = np.delete(rescaledztot, indcut)[0:-1]
            
            # Example data (x and y)
            
            x_data2 = np.arange(len(zi2[np.argmax(zi2): -1]))[1:-1]
            y_data2 = zi2[np.argmax(zi2): -1][1:-1]
            # Define the model function: b / sqrt(a - x)

            def model_func(x,a,  d):
                return (np.sum(y_data[-10:-1]))/len(y_data[-10:-1])*(np.sum(x_data[-10:-1])/len(y_data[-10:-1])-a)**d/(x-a)**d

            # Initial guesses for parameters a,b,c,d
            
            if i ==3:

                initial_guess = [0,1/2]
                bounds = ([ 0,0], [ 1, 5]) 
            else:

                initial_guess = [0,d_fitted]
                bounds = ([0,0], [ 1,5]) 

            params, covariance = curve_fit(model_func, x_data, y_data, p0=initial_guess, bounds=bounds)
            params3, covariance3 = curve_fit(model_func, x_data2, y_data2, p0=initial_guess, bounds=bounds)
            def model_func2(x,a,  b):
                return b*(t[1])/(x-a)**0.5
            if i ==3:
                initial_guess = [0, b_fitted]
                bounds = ([ -.1,0], [ .1,np.inf]) 
        #         bounds = ([a_fitted-1, b_fitted-1,c_fitted-1,d_fitted-1], [a_fitted+1, b_fitted+1, c_fitted+1,d_fitted+1])  
            else:
                initial_guess = [0, b_fitted]
                bounds = ([-.1,0], [ .1,np.inf]) 
                
            params2, covariance2 = curve_fit(model_func2, x_data, y_data, p0=initial_guess, bounds=bounds)
            
            # Extract the fitted parameters a and b
            a_fitted,d_fitted = params
            a_fitted2, b_fitted2 = params2
            a_fitted3,d_fitted3 = params3
            listex = listex+[d_fitted]

            # Generate fitted y values for plotting
            x_fine = np.linspace(min(x_data), max(x_data), 100)
#             y_fitted = model_func(x_fine,a_fitted, b_fitted,d_fitted)
            y_fitted = model_func(x_fine,a_fitted,d_fitted)

            # Plot the original data and the fitted curve
            plt.scatter(x_data, y_data, label=f'Data, {i}')
            plt.scatter(x_data2, y_data2, label=f'Data, {i}',s=.1)
            plt.scatter((np.sum(x_data[0:3]))/3,(np.sum(y_data[0:3]))/3,  c = 'black')
            plt.scatter(x_data[0:2], y_data[0:2], c = 'green', s =50)
            plt.plot(x_fine-a_fitted2, model_func2(x_fine,a_fitted2, b_fitted2))
            plt.plot(x_fine-a_fitted3, model_func(x_fine,a_fitted3,  d_fitted3),label=f'Fit:{d_fitted3:.3f}')
            plt.plot(x_fine-a_fitted, y_fitted, label=f'Fit:{d_fitted:.3f}', color='red')
            plt.xlabel('x')
            plt.ylabel('y')
            plt.legend()

            plt.title('Fit of b/(x)^d +c to Data')

            plt.show()
    return listex,listshift,listproduct



```
